```python
import opvious

client = opvious.Client.from_environment()
```

```python
inputs = await client.assemble_inputs(
    formulation_name='shift-scheduling',
    parameters={
        'horizon': 21,
        'resource': {'open': 3, 'close': 2},
        'level': {chr(65+i): i for i in range(7)}, # A, B, C, ...
    },
    infer_dimensions=True
)
```

```python
attempt = await client.start_attempt(inputs)
await client.wait_for_outcome(attempt)
```

```python
schedule = await client.fetch_variable(attempt, 'schedule')
schedule.reset_index().pivot(index=['day'], columns=['employee'], values=['shift']).fillna('')
```
